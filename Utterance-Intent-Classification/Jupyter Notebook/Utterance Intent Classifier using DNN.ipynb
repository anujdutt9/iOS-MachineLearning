```python
# Import Dependencies
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# Import intents file
import json

with open('lemmatizedDataset.json') as json_data:
    intents = json.load(json_data)
```

```python
# Initialize Variables
# Number of Intents
classes = len(intents['intents'])

# Current Intent
currentClass = 0

# Train Features and Labels
X_train = []
y_train = []
```

```python
print('Total Intents: ', classes)
```

```python
# For all intents in intent
for intent in intents['intents']:
    # Initialize intent array
    y = [0] * classes
    # Set current intent index to "1"
    y[currentClass] = 1
    
    # Get all embeddings corresponding to intents
    for embedded in intent['embeddedUtterances']:
        X_train.append(embedded)
        y_train.append(y)

    currentClass += 1
```

```python
# Check shape of Training features and labels
len(X_train), len(y_train)
```

```python
# Number of Words
words = len(X_train[0])

print(classes, "classes")
print(words, "words")

print(len(X_train), len(X_train[0]), "X_train")
print(len(y_train), len(y_train[0]), "y_train")
```

```python
# Define Model Architecture
model = Sequential()
model.add(Dense(256, input_dim=len(X_train[0]), activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(len(y_train[0]), activation='softmax'))
model.summary()
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the Model
model.fit(np.array(X_train), np.array(y_train), epochs=400)
```

```python
# Save Trained Model
model.save('NaturalLanguageClassifier.h5')
```

```python
# Convert model to CoreML Model
import coremltools

coreml_model = coremltools.converters.keras.convert(model, input_names="embeddings", output_names="entities")
coreml_model
```
